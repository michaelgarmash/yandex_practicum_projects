## Подготовка данных

### Приведение заголовков столбцов к нижнему регистру

```python
data.columns = map(str.lower, data.columns) 
```

### Обработка пропусков
находим столбцы с пропусками

```python
data.isna().sum()
```

```python
msno.bar(data);
```

Посмотрим строки с пропусками в названии.

```python
data[data.name.isna()]
```

В этих  же строках пропущены и жанры. Придется их удалить.

```python
data.dropna(subset='name', inplace=True)
```

также придется удалить 269 строк с пропущенным годом выпуска. Это 1.6% от датасета - в пределах допуска.

```python
data = data.dropna(subset='year_of_release').reset_index(drop=True)
```

Пропуски в столбцах с оценками пользователей, критиков и ESRB трогать не будем т.к., скорее всего, они обусловлены объективными причинами (не каждой игре выставляли оценки и присваивали рейтинг).

Посмотрим содержимое столбцов с оценками пользователей и рейтингом ESRB.

```python
display(data.user_score.sort_values(ascending=True).unique())
display(data.rating.sort_values(ascending=True).unique())
```

В столбце пользовательских оценок, кроме собственно оценок и пропусков, видим  некую аббревиатуру "tbd", которая расшифровывается, как "To Be Determined", то есть "Будет определено". Заменим ее на NaN.

В столбце с рейтингом есть устаревшая аббревиатура "К-А", вместо которой сейчас используют "Е". Произведем замену.  Также, заполним пустые значения аббревиатурой "RP", которая обозначает рейтинг в ожидании или заполнитель. Он появляется рядом с рекламой игр, которые еще не были оценены. https://okdk.ru/chto-oznachajut-rejtingi-videoigr-rukovodstvo-po-esrb-i-pegi/

```python
data.user_score = data.user_score.replace('tbd', np.nan)
data.rating = data.rating.fillna('RP')
data.rating = data.rating.replace('K-A', 'E')
```

###  Изменение типов данных

```python
data.year_of_release = data.year_of_release.astype('int')
data.user_score = data.user_score.astype('float64')
```

### суммарные продажи во всех регионах
Добавим в датасет столбец со значениями суммарных продаж по всем представленным регионам:

```python
data['world_sales'] = data.na_sales + data.eu_sales + data.jp_sales + data.other_sales
```

При этом из датасета было изъято не более 1.6% данных. В столбце rating отсутствующие значения были заменены на стандартную аббревиатуру, используемую ESRB в соответствии с мануалом. Пропуски в столбцах оценок игр критиками/пользователями трогать не стали, предположив, что эти игры не были оценены по каким-либо объективным причинам, например, из-за их малой популярности или региональных особенностей, либо в годы их выпуска не существовало такого сервиса.
- Данные в столбце year_of_release были преобразованы в целочисленный тип int32.
- В столбце user_score было обнаружено строковое значение "tbd", которое расшифровывается, как "To Be Determined", то есть "Будет определено". Эта аббревиатура используется для заполнения ячеек тех игр, которые не получили ни одной пользовательской оценки. Заменили ее на NaN. После этого данные преобразовали в формат с пл. точкой.

## Исследовательский анализ данных

### Выпуск игр по годам

```python
# Таблица с данными по продажам и релизам, сгруппированным по годам.
t = data.pivot_table(
    index='year_of_release',
    aggfunc={
        'name' : 'count',
        'world_sales' : 'sum'
    })
t.plot(grid=True, figsize=(15, 7))
plt.title('Продажи и релизы по годам')
plt.xlabel('Год релиза')
plt.ylabel('Количество релизов/продаж')
plt.legend(['release', 'sales'])
plt.show()
```

Рассмотрим график релизов/продаж по годам. Видно, что, до 1994 года, игр выпускается очень мало. Этот период точно можно отбросить.

```python
dt_94 = data.query('year_of_release > 1993')
```

### Продажи по платформам
Посмотрим, как менялись продажи по платформам. Найдем платформы с наибольшими суммарными продажами и построим распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?

```python
# Таблица топ-15 платформ с самыми высокими продажами
top15= dt_94.groupby('platform')['world_sales'].sum().sort_values(ascending=False).head(15).reset_index()
# Данные отобранные из основной таблицы для платформ из топ-15 для построения графика
top15_sales = dt_94.query('platform in @top15.platform').groupby(['platform', 'year_of_release'])['world_sales'].sum().unstack('platform')
```

```python
top15_sales.plot(grid=True, figsize=(15,5))
plt.title('Продажи по платформам')
plt.xlabel('Годы')
plt.ylabel('Количество продаж')
plt.show()
```

По графику видно, что расцвет приставок пришелся на начало 2000-х. После 2005-го года количество платформ значительно увеличилось. Пик продаж пришелся на 2008 год. Затем началось падение продаж у всех игроков, которое продолжалось до конца представленного данными периода.

Построим таблицу с временем жизни платформ. И посмотрим на статистику:

```python
plt_years = data.pivot_table(index='platform', values='year_of_release', aggfunc=('max','min')).reset_index()
plt_years['lifetime'] = plt_years['max'] - plt_years['min'] + 1
```

```python
plt_years.lifetime.describe(percentiles=[0.05, 1/4, 1/2, 3/4, 0.95, 0.99])
```

```python
plt_years[plt_years['lifetime'] > 21]
```

Медианнное время жизни платформы - 7 лет.  Среднее скошено верх из-за парочки долгожителей и составляет 9 лет. На основании этого можно утверждать, что новые платформы появляются, а старые исчезают за срок в среднем от семи до девяти лет.

## Статистический анализ данных
Проверка гипотез

### Гипотеза о равенстве средних  рейтингов платформ
- Гипотеза H0 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Гипотеза Н1 - Средние пользовательские рейтинги платформ Xbox One и PC разные.
- критический уровень статистической значимости = 0.05

```python
array1 = data_2013.query('platform == "XOne" and user_score.notna()')['user_score']
array2 = data_2013.query('platform == "PC" and user_score.notna()')['user_score']
alpha = 0.05
results = st.ttest_ind(array1, array2)
print('р-значение', results.pvalue)
if results.pvalue < alpha:
    print('отвергаем нулевую гипотезу')
else:
    print('не отвергаем нулевую гипотезу')
array1.hist(bins=10)
array2.hist(bins=10);
```

### Гипотеза о равенстве средних  рейтингов жанров
- Гипотеза H0 - Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые;
- Гипотеза Н1 - Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
- критический уровень статистической значимости = 0.05

```python
array1 = data_2013.query('genre == "Action" and user_score.notna()')['user_score']
array2 = data_2013.query('genre == "Sports" and user_score.notna()')['user_score']
alpha = 0.05
results = st.ttest_ind(array1, array2)
print('р-значение', results.pvalue)
if results.pvalue < alpha:
    print('отвергаем нулевую гипотезу')
else:
    print('не отвергаем нулевую гипотезу')
array1.hist(bins=10)
array2.hist(bins=10);
```

### Вывод

Нулевая гипотеза должна быть сформулирована таким образом, чтобы в формуле использовался знак равенства. Поэтому в обоих случаях были приняты утверждения о том, что сравниваемые параметры равны.
Для проверки гипотез был применен метод сравнения средних  двух  генеральных совокупностей т. к. в обоих случаях две ГС не зависят друг от друга, выборки имеют нормальное распределение.

Обе гипотезы получили свое подтверждение:
Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
