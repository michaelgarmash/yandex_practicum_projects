# Недвижимость в СПБ и его окрестностях
## Описание проекта

В нашем распоряжении данные сервиса Яндекс Недвижимость — архив объявлений за несколько лет о продаже квартир в Санкт-Петербурге и соседних населённых пунктах.

**Задача** — выполнить предобработку данных и изучить их, чтобы найти интересные особенности и зависимости, которые существуют на рынке недвижимости.

## Описание данных

О каждой квартире в базе содержится два типа данных: добавленные пользователем и картографические. Например, к первому типу относятся площадь квартиры, её этаж и количество балконов, ко второму — расстояния до центра города, аэропорта и ближайшего парка.

`airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)  
`balcony` — число балконов  
`ceiling_height` — высота потолков (м)  
`cityCenters_nearest` — расстояние до центра города (м)  
`days_exposition` — сколько дней было размещено объявление (от публикации до снятия)  
`first_day_exposition` — дата публикации  
`floor` — этаж  
`floors_total` — всего этажей в доме  
`is_apartment` — апартаменты (булев тип)  
`kitchen_area` — площадь кухни в квадратных метрах (м²)  
`last_price` — цена на момент снятия с публикации  
`living_area` — жилая площадь в квадратных метрах (м²)  
`locality_name` — название населённого пункта  
`open_plan` — свободная планировка (булев тип)  
`parks_around3000` — число парков в радиусе 3 км  
`parks_nearest` — расстояние до ближайшего парка (м)  
`ponds_around3000` — число водоёмов в радиусе 3 км  
`ponds_nearest` — расстояние до ближайшего водоёма (м)  
`rooms` — число комнат  
`studio` — квартира-студия (булев тип)  
`total_area` — общая площадь квартиры в квадратных метрах (м²)  
`total_images` — число фотографий квартиры в объявлении

```python
data.loc[data['ceiling_height']>=20, 'ceiling_height'] /= 10
```

в столбце ceiling_height отсутствует почти 40% значений. Заполним их медианными значениями по каждому населенному пункту.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby('locality_name')['ceiling_height'].transform('median'))
data['ceiling_height'].isna().sum()
```

Выяснилось, что незаполнены еще 156 значений по тем н.п., где отсутствующие значения составляли 100% выборки. Их заполним по нормам СНиП 31-01-2003 для ленинградской области, который утверждает, что высота потолков д.б. не менее 2.5 метров

```python
data['ceiling_height'] = data['ceiling_height'].fillna(2.5)
```

Он равен отношению жилой площади дома к его общему метражу и обычно находится в рекомендованном промежутке от 0,5 до 0,7. В комфортабельных квартирах он ближе к нижней границе, в более экономичных – к верхней. Можно заполнить этот столбец значениями общей площади из столбца total_area и взять среднее значение этого коэффициента 0.6

Однако из-за особого статуса в таком помещении нельзя оформить постоянную регистрацию – только временную. В результате имеем нечто среднее между гостиничным номером, офисом и квартирой.
Вероятно, что владельцы апартаментов точно знают тип своей недвижимости и заполнили этот пункт правильно. А остальные, видимо, просто отмахнулись от непонятного названия. Поэтому заполним отсутствующие ячейки значениями False

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

В столбце kitchen_area пропуск составляет 9.6%. Найдем студии и опенспэйс с отсутствующим значением и поместим туда ноль. для остальных пропусков считаем, какая медианная доля технической площади - кухня, и назначаем для пропусков такое значение.

```python
data['kitchen_area'].isna().sum()
```

```python
data.loc[data['studio'] == True, 'kitchen_area'] = 0
data.loc[data['open_plan'] == True, 'kitchen_area'] = 0
data['kitchen_area'].isna().sum()
```

```python
for i in range(1, data['rooms'].max()):
    ratio = (data[data['rooms'] == i]['kitchen_area'] / 
             (data[data['rooms'] == i]['total_area'] - data[data['rooms'] == i]['living_area'])).median()
    data['kitchen_area'] = data['kitchen_area'].fillna(value = (
        data[data['rooms'] == i]['total_area'] - data[data['rooms'] == i]['living_area']) * ratio)
data['kitchen_area'].isna().sum()
```

```python
area_check = data.total_area - data.kitchen_area - data.living_area
print(data.loc[area_check <= 0]['total_area'].count())
print('аномальная средняя общая площадь', data.loc[area_check <= 0]['total_area'].mean())
print('аномальная средняя жилая площадь', data.loc[area_check <= 0]['living_area'].mean())
print('аномальная средняя площадь кухни', data.loc[area_check <= 0]['kitchen_area'].mean())
print()
print('средняя общая площадь', data['total_area'].mean())
print('средняя жилая площадь', data['living_area'].mean())
print('средняя площадь кухни', data['kitchen_area'].mean())
```

Появилось 150 строк, где общая площадь стала меньше суммы жилой и кухни. При их визуальном анализе видно, что в этих объявлениях жилая площадь указана очень большой по отношению к общей по сравнению со статистикой всей выборки. Скорее всего, это ошибки ввода данных. Такие строки придется удалить.

```python
data = data.loc[area_check > 0]
```

Следующий столбец с пропусками - balcony. Здесь их 48.6%. Предположим, что если человек не указал количество балконов, то они у него отсутствуют. Пропущенные данные по столбцам с количеством объектов в радиусе 3 км. заполняем нулями по той же причине

```python
data['balcony'].fillna(0, inplace=True)
data['ponds_around3000'].fillna(0, inplace=True)
data['parks_around3000'].fillna(0, inplace=True)
```

### Изменение типов данных

Меняем тип данных в столбце first_day_exposition на datetime т. к. в нем хранятся сведения о дате. Значения в столбцах floors_total и balcony (количество этажей и балконов) могут быть только целочисленными.

Причины этих пропусков возможно кроются в неправильном формате адреса или его отсутствии, вследствие чего алгоритмы яндекса не смогли его обработать. Также, возможно, что при пустом значении в столбцах ponds_around3000 и parks_around3000, расстояние до ближайшего такого объекта не вычисляется.
Пропуски в столбце days_exposition скорее всего кроются в том, что эти объявления все еще активны.

### Добавление столбцов в таблицу

```python
# цена одного квадратного метра
data['one_m_price'] = data.last_price / data.total_area
# день недели публикации объявления (0 — понедельник, 1 — вторник и так далее)
data['weekday_publish'] = pd.DatetimeIndex(data.first_day_exposition).weekday
# месяц публикации объявления
data['month_publish'] = pd.DatetimeIndex(data.first_day_exposition).month
# год публикации объявления
data['year_publish'] = pd.DatetimeIndex(data.first_day_exposition).year
# тип этажа квартиры
Conditions = [
    (data['floor'] == data['floors_total']),
    (data['floor'] == 1),
    (data['floor'] != data['floors_total']) & (data['floor'] != 1)]
Categories = ['последний', 'первый', 'другой']
data['floor_type'] = np.select(Conditions, Categories)
# расстояние до центра города в километрах
data['citycenters_km'] = data['city_centers_nearest'].div(1000).round()
# Изменим значения столбца last_price на цены, выраженные в тысячах рублей для упрощения дальнейшей работы
data['last_price'] = data['last_price'] / 1000
# переименуем столбец, чтобы не забыть, что цены в тысячах
data = data.rename(columns={'last_price' : 'price_1000'})
data.shape
```

```python
data.groupby('weekday_publish')['price_1000'].mean().plot(x='weekday_publish', y='price_1000', kind='bar', grid=True, figsize=(15,5))
plt.title('Цена/день публикации')
plt.show()

data.groupby('month_publish')['price_1000'].mean().plot(x='month_publish', y='price_1000', kind='bar', grid=True, figsize=(15,5))
plt.title('Цена/месяц публикации')
plt.show()

data.groupby('year_publish')['price_1000'].mean().plot(x='year_publish', y='price_1000', kind='bar', grid=True, figsize=(15,5))
plt.title('Цена/год публикации')
plt.show()
```
