# Банк - сегментация клиентов, отток.
## Описание проекта
Проект посвящен сегментированию пользователей банка. Нужно максимально эффективно выделить сегменты отточных клиентов, чтобы они были однородными, высокоотточными и покрывали максимум проблемных зон банка.

Заказчик - менеджер продукта, который разрабатывает меры по удержанию клиентов и мероприятия для работы с целевой аудиторией. Пользователем результата по проекту будет менеджер продукта. Цель - определить, на какие основные сегменты делятся клиенты банка и выявить характеристики каждого сегмента.

## Задачи
Провести исследовательский анализ данных
Сегментировать пользователей на основе данных о количестве потребляемых продуктов
Сформулировать и проверить статистические гипотезы:
- гипотеза различия дохода между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.
- гипотеза различия среднего возраста между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.

**Цель исследования - помочь менеджерам выработать воздействие, которое уменьшит отток клиентов из банка.**

Подготовить презентацию
## Описание данных
Датасет содержит данные о клиентах банка «Метанпром». Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск.

### Изучение групп клиентов

#### кредитный рейтинг

Выделим из датасета оставшихся и ушедших клиентов и построим гистограмму распределения по кредитному рейтингу для всех трех групп.

```python
clients = df.query('churn==0') # оставшиеся клиенты 
churned = df.query('churn==1') # ушедшие клиенты
d = pd.DataFrame()
d['total'] = df.score.describe(percentiles=[0.05, 1/4, 1/2, 3/4, 0.95, 0.99])
d['clients'] = clients.score.describe(percentiles=[0.05, 1/4, 1/2, 3/4, 0.95, 0.99])
d['churned'] = churned.score.describe(percentiles=[0.05, 1/4, 1/2, 3/4, 0.95, 0.99])
display(d)
print(f"Корреляция между оттоком и рейтингом = {df.churn.corr(df.score):.2f}")
```

```python
# функция для отрисовки диаграмм по столбцу 'column'. 'label' - заголовок графика (соотв. названию столбца)
def pie_chart(column, label):

    plt.figure(figsize=(15, 5))
    colors = sns.color_palette('pastel')[ 0:5 ]

    plt.subplot(1, 3, 1)
    t = df.groupby(column).userid.count()
    plt.pie(t, labels=t.index, colors=colors, autopct='%1.1f%%')
    plt.title('всего '+ label, size=14)

    plt.subplot(1, 3, 2)
    t = clients.groupby(column).userid.count()
    plt.pie(t, labels=t.index, colors=colors, autopct='%1.1f%%')
    plt.title('клиенты ' + label, size=14)

    plt.subplot(1, 3, 3)
    t = churned.groupby(column).userid.count()
    plt.pie(t, labels=t.index, colors=colors, autopct='%1.1f%%')
    plt.title('отток ' + label, size=14);

pie_chart('city', 'по городам')
```

Сделаем несколько визуализаций для лучшего понимания различий:

```python
"""
функция отрисовки гистограмм по результатам данных сгруппированной таблицы. 
Выдает четыре графика распределения значений столбца, квантили и среднее.
data - датафрейм, отортированный по продуктам;
column - рассматриваемый столбец датафрейма data;
nazv -  название столбца для подписи оси X
"""
def histo(data, column, nazv):
    plt.figure(figsize=(15, 8))
    for i in data.products.unique():
        plt.subplot(2,2,i)
        t =  data.query('products==@i')
        col = t[column]
        sns.histplot(x=col, data=data, bins=50, kde=True, hue='churn')
        
        plt.axvline(x=col.mean(), color='red', linestyle='-', label='среднее')
        plt.axvline(x=col.median(), color='green', linestyle='--')
        plt.axvline(x=col.quantile(0.25), color='green', linestyle='--', label='квартили')
        plt.axvline(x=col.quantile(0.75), color='green', linestyle='--')
        plt.title(str(i)+'-й сегмент', fontsize = 18)
        plt.xlabel(nazv, size=14)
        plt.ylabel('количество клиентов', size=14)
        plt.legend(fontsize=12)
        for pos in ['right', 'top']:
            plt.gca().spines[pos].set_visible(False)
        plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)
    plt.show()
 
```

```python
print("Гистограммы распределения кредитного рейтинга клиентов по сегментам в разрезе оттока")
histo(df, 'score', 'кредитный рейтинг')
```

## проверка статистических гипотез


### Гипотеза 1
гипотеза различия среднего дохода между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.


- H0 = средний доход клиентов, которые пользуются одним и двумя продуктами, одинаковый.
- H1 = средний доход клиентов, которые пользуются одним продуктом отличается от среднего дохода тех, кто пользуется двумя.

Из таблицы 6.4.1 нам известны размеры выборок: 1 продукт = 5084, 2 продукта = 4590.

создадим таблицы:

```python
one_prod_revenue = df.query('products == 1').estimated_salary
two_prod_revenue = df.query('products == 2').estimated_salary
```

Выборки достаточно большие и одинаковые по размеру. Оснований думать, что дисперсии ГС разные, нет.  Проведем тест о равенстве среднего двух генеральных совокупностей:

```python
alpha = 0.05 #критический уровень статистической значимости

results = st.ttest_ind(one_prod_revenue, two_prod_revenue) 

print('p-значение:', results.pvalue) 

if results.pvalue < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Средний доход клиентов, которые пользуются одним и двумя продуктами, одинаковый.

### Гипотеза 2
- гипотеза различия среднего возраста между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.


- H0 = средний возраст клиентов, которые пользуются одним и двумя продуктами одинаковый.
- H1 = средний возраст клиентов, которые пользуются одним продуктом отличается от дохода тех, кто пользуется двумя.

```python
one_prod_age = df.query('products == 1').age
two_prod_age = df.query('products == 2').age
```

```python
alpha = 0.05 #критический уровень статистической значимости

results = st.ttest_ind(one_prod_age, two_prod_age) 

print('p-значение:', results.pvalue) 

if results.pvalue < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

В среднем возрасте клиентов, которые пользуются одним и двумя продуктами, есть отличия.
